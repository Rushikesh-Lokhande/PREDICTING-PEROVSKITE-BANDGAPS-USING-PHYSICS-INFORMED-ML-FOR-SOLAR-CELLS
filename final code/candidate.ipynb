```python
import pandas as pd
import numpy as np
import itertools

# 1. DATABASE OF ATOMIC PROPERTIES
# These values are standard constants found in Physics/Chemistry Handbooks
a_props = {
    'Cs': {'r': 1.88, 'chi': 0.79, 'EA': 0.47, 'mass': 132.9},
    'Rb': {'r': 1.72, 'chi': 0.82, 'EA': 0.48, 'mass': 85.4},
    'K':  {'r': 1.64, 'chi': 0.82, 'EA': 0.50, 'mass': 39.1},
    'Na': {'r': 1.39, 'chi': 0.93, 'EA': 0.54, 'mass': 22.9},
    'MA': {'r': 2.17, 'chi': 2.55, 'EA': 0.10, 'mass': 32.0}, # Methylammonium
    'FA': {'r': 2.53, 'chi': 2.57, 'EA': 0.10, 'mass': 45.0}, # Formamidinium
    'Ba': {'r': 1.35, 'chi': 0.89, 'EA': 0.14, 'mass': 137.3}
}

b_props = {
    'Pb': {'r': 1.19, 'chi': 2.33, 'EA': 0.36, 'mass': 207.2},
    'Sn': {'r': 1.10, 'chi': 1.96, 'EA': 1.11, 'mass': 118.7},
    'Ge': {'r': 0.73, 'chi': 2.01, 'EA': 1.23, 'mass': 72.6},
    'Ti': {'r': 0.60, 'chi': 1.54, 'EA': 0.08, 'mass': 47.8},
    'Zn': {'r': 0.74, 'chi': 1.65, 'EA': 0.00, 'mass': 65.3},
    'Mg': {'r': 0.72, 'chi': 1.31, 'EA': 0.00, 'mass': 24.3},
    'Sr': {'r': 1.18, 'chi': 0.95, 'EA': 0.05, 'mass': 87.6},
    'Ca': {'r': 1.00, 'chi': 1.00, 'EA': 0.02, 'mass': 40.0},
    'Mn': {'r': 0.67, 'chi': 1.55, 'EA': 0.00, 'mass': 54.9},
    'Fe': {'r': 0.64, 'chi': 1.83, 'EA': 0.15, 'mass': 55.8}
}

x_props = {
    'I':  {'r': 2.20, 'chi': 2.66, 'EA': 3.05, 'mass': 126.9},
    'Br': {'r': 1.96, 'chi': 2.96, 'EA': 3.36, 'mass': 79.9},
    'Cl': {'r': 1.81, 'chi': 3.16, 'EA': 3.61, 'mass': 35.4},
    'F':  {'r': 1.33, 'chi': 3.98, 'EA': 3.40, 'mass': 18.9},
    'O':  {'r': 1.40, 'chi': 3.44, 'EA': 1.46, 'mass': 16.0}
}

# 2. GENERATE ALL 350 COMBINATIONS (Combinatorial Product)
combinations = list(itertools.product(a_props.keys(), b_props.keys(), x_props.keys()))
vhts_data = []

for a, b, x in combinations:
    # Get raw values
    ra, rb, rx = a_props[a]['r'], b_props[b]['r'], x_props[x]['r']
    chia, chib, chix = a_props[a]['chi'], b_props[b]['chi'], x_props[x]['chi']
    eaa, eab, eax = a_props[a]['EA'], b_props[b]['EA'], x_props[x]['EA']
    ma, mb, mx = a_props[a]['mass'], b_props[b]['mass'], x_props[x]['mass']
    
    # 3. APPLY PHYSICS FORMULAS (The 24 Features)
    t = (ra + rx) / (np.sqrt(2) * (rb + rx)) # Tolerance Factor
    octa = rb / rx                          # Octahedral Factor
    
    # Structural features
    vol = (2 * (rb + rx))**3                # Cubic Volume Approximation
    dens = (ma + mb + 3*mx) / (vol * 0.6022) # Density Calculation
    
    row = {
        'formula': f"{a}{b}{x}3", 'A_site': a, 'B_site': b, 'X_site': x,
        'rA': ra, 'rB': rb, 'rX': rx,
        'chi_A': chia, 'chi_B': chib, 'chi_X': chix,
        'EA_A': eaa, 'EA_B': eab, 'EA_X': eax,
        'tolerance_factor': t,
        'octahedral_factor': octa,
        'volume': vol,
        'density': dens,
        'chi_diff_BX': abs(chib - chix),
        'r_ratio_BX': rb / rx,
        'bond_len_BX': rb + rx,
        'inv_bond_sq': 1 / ((rb + rx)**2),
        'stability_index': t * octa,
        'lattice_strain': 0.0,    # Default for ideal cubic
        'packing_index': 0.6,     # Ideal packing fraction
        'PC1_Size': 0.0,          # Placeholder (PCA required for training)
        'PC2_Shape': 0.0,         # Placeholder
        'Stability_Tag': 1 if (0.8 <= t <= 1.05 and 0.4 <= octa <= 0.7) else 0,
        'B_site_encoded': 0       # Placeholder for encoding
    }
    vhts_data.append(row)

# 4. SAVE TO CSV
vhts_df = pd.DataFrame(vhts_data)
vhts_df.to_csv('vscode-Virtual_Perovskite_Library.csv', index=False)

print(f"🎉 Successfully generated {len(vhts_df)} hypothetical perovskites in Virtual_Perovskite_Library.csv")
```
