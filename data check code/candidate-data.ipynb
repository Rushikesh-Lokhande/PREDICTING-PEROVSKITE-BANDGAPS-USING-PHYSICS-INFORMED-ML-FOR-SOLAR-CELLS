```python
##fraction of candidates with "Cs" in A-site so we will not use this data for training
# code for 1000 candidates with random mixing ratios
import random

A_site = ["Cs", "Rb", "K"]
B_site = ["Pb", "Sn", "Ge"]
X_site = ["Cl", "Br", "I"]

candidates = set()

while len(candidates) < 1000:
    
    A = random.choice(A_site)
    B = random.choice(B_site)
    
    X1, X2 = random.sample(X_site, 2)
    
    # Random mixing ratio
    ratio = round(random.uniform(0.1, 0.9), 2)
    
    formula = f"{A}{B}({X1}{ratio}{X2}{1-ratio})3"
    
    candidates.add(formula)

candidates = list(candidates)

print("Generated candidates:", len(candidates))
candidates[:10]
```

```python
import pandas as pd

candidate_df = pd.DataFrame(candidates, columns=["composition"])

candidate_df.head()

##fraction of candidates with "Cs" in A-site so we will not use this data for training
```

```python
A_site = ["Cs", "Rb", "K", "Na"]
B_site = ["Pb", "Sn", "Ge", "Bi"]
X_site = ["Cl", "Br", "I"]

candidates = []

for A in A_site:
    for B in B_site:
        for X in X_site:
            candidates.append(A + B + X + "3")

candidate_df = pd.DataFrame(candidates, columns=["composition"])

print("Total candidates:", len(candidate_df))
```

```python
candidate_df.tail()
```

```python
from pymatgen.core import Composition
from matminer.featurizers.composition import ElementProperty

# Convert to pymatgen format
candidate_df["composition_obj"] = candidate_df["composition"].apply(Composition)

# Use Magpie features
featurizer = ElementProperty.from_preset("magpie")

candidate_df = featurizer.featurize_dataframe(candidate_df, "composition_obj")

print("Feature dataset ready")
candidate_df.head()
```

```python
candidate_df.to_csv("candidate_features.csv", index=False)

print("Candidate feature dataset saved successfully")
```

```python
selected_features = [
    "composition",
    "MagpieData mean Electronegativity",
    "MagpieData mean AtomicWeight",
    "MagpieData mean CovalentRadius",
    "MagpieData mean NValence",
    "MagpieData mean MeltingT"
]

reduced_df = candidate_df[selected_features].copy()

reduced_df.head()
```

```python
import math
from pymatgen.core import Composition, Element

def compute_tolerance_factor(formula):
    try:
        comp = Composition(formula)
        elems = list(comp.elements)
        
        if len(elems) != 3:
            return None
        
        # crude assignment (A,B,X)
        A, B, X = elems
        
        rA = Element(str(A)).atomic_radius or 1
        rB = Element(str(B)).atomic_radius or 1
        rX = Element(str(X)).atomic_radius or 1
        
        t = (rA + rX) / (math.sqrt(2) * (rB + rX))
        return t
    
    except:
        return None

reduced_df["tolerance_factor"] = reduced_df["composition"].apply(compute_tolerance_factor)
```

```python
reduced_df = reduced_df.rename(columns={
    "MagpieData mean Electronegativity": "electronegativity_mean",
    "MagpieData mean AtomicWeight": "atomic_weight_mean",
    "MagpieData mean CovalentRadius": "atomic_radius_mean",
    "MagpieData mean NValence": "valence_electrons_mean",
    "MagpieData mean MeltingT": "melting_temp_mean",
    "tolerance_factor": "tolerance_factor"
})

reduced_df.head()
```

```python
reduced_df.to_csv("candidate_clean_features.csv", index=False)

print("Clean dataset saved")
```
